## 1. Load Data House Pricing

```python
# Load Data Test
df_test = pd.read_csv("./test.csv")
df_train = pd.read_csv("./train.csv")
```

```python
pd.options.display.max_columns = None
df_train.head()
```

```python
df_train.describe()
```

```python
df_train.isnull().sum()
```

```python
# Mengisi data hilang dengan median
df_train.fillna(df_train.median(), inplace=True)
```

## 2. Explore and Visualize Data

```python
print("Total Rows Train    : ", df_train.shape[0])
print("Total Columns Train : ", df_train.shape[1])
print("Total Rows Test     : ", df_test.shape[0])
print("Total Columns test  : ", df_test.shape[1])
```

## 3. Preprocessing Data

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
```

```python
# Mengubah fitur objek menjadi numerik
le = LabelEncoder()
categorical_features = df_train.select_dtypes(include=['object']).columns
for feature in categorical_features:
    df_train[feature] = le.fit_transform(df_train[feature])
```

```python
df_train2 = df_train.select_dtypes(object).columns.tolist()
```

```python
# remove categorical data (object)
# filter kolom yang akan digunakan
df_train_drop = df_train.drop(columns=df_train2)
```

```python
# Memisahkan fitur dan target
x = df_train_drop.drop('SalePrice', axis=1)
y = df_train_drop['SalePrice']
```

```python
# Membagi data menjadi training set dan testing set
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Membangun model RandomForestRegressor
model = RandomForestRegressor()
model.fit(x_train, y_train)
```

## 4. Model Evaluation

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Melakukan prediksi pada set pengujian
y_pred = model.predict(x_test)
```

```python
# Mengukur performa model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print(f"Mean Squared Error   : {round(mse,2)}")
print(f"S Mean Squared Error : {round(mse,2)}")
print(f"R2 Score             : {round(r2,3)}")
```
